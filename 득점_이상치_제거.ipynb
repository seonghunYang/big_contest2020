```
#득점
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics, optimizers

model_h = Sequential()
model_h.add(Dense(64, input_shape=(14, ), kernel_initializer='he_normal', activation='relu'))
model_h.add(Dropout(0.5))  
model_h.add(Dense(32, kernel_initializer='he_normal',activation = 'relu'))
model_h.add(Dropout(0.5))  
model_h.add(Dense(8, kernel_initializer='he_normal',activation = 'relu'))
model_h.add(Dropout(0.5))  
model_h.add(Dense(1,activation = 'relu'))

adam = optimizers.Adam(lr = 0.01) 

model_h.compile(loss='mse', optimizer = 'adam', metrics=['mse'])

from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=20)
model_h.fit(x_train_h, y_train_h, validation_split=0.2, verbose = 1,
          batch_size=1, epochs=100, callbacks = [early_stopping])

loss, accuracy = model_h.evaluate(x_test_h, y_test_h, batch_size=1)
print('lose: ', loss)
print('mse: ', accuracy)

y_pred = model_h.predict(x_test_h)
```
