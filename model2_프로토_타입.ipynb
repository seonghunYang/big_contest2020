```
game = game[columns]
```

```
y= game['결과'].copy()
del game['결과']
```

```
x= game.copy()
```

```
x1 = x.values
y1 = y.values
```

```
x1.shape
y1.shape
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x1, y1, test_size = 0.3 ,shuffle=False
)
```

```
x_train.shape
```

```
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
print(standardScaler.fit(x_train))
x_train = standardScaler.transform(x_train)
```

```
x_train
```

```
standardScaler = StandardScaler()
print(standardScaler.fit(x_test))
x_test = standardScaler.transform(x_test)
```

```
x_test
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras import metrics
model = Sequential()
model.add(Dense(128, input_shape=(402, )))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=[metrics.binary_accuracy])

from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=20)
model.fit(x_train, y_train, validation_split=0.2, verbose = 1,
          batch_size=1, epochs=100, callbacks = [early_stopping])

loss, accuracy = model.evaluate(x_test, y_test, batch_size=1)
print('lose: ', loss)
print('mse: ', accuracy)

y_pred = model.predict(x_test)
```

```
for idx, i in enumerate(y_pred):
  if i >= 0.5:
    y_pred[idx] = 1
  else:
    y_pred[idx] = 0
```

```
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, auc
confusion_matrix = confusion_matrix(y_test,y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test,y_pred)
print(confusion_matrix)
print('recall:' + str(recall) +", precision: " + str(precision) + 'F1-score: ' + str(f1))
roc_auc = roc_auc_score(y_test,y_pred)
roc_auc
```

```
import plotly.express as px

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
fig = px.area(
    x=fpr, y=tpr,
    title=f'ROC Curve (AUC={auc(fpr, tpr):.4f})',
    labels=dict(x='False Positive Rate', y='True Positive Rate'),
    width=700, height=500
)
fig.add_shape(
    type='line', line=dict(dash='dash'),
    x0=0, x1=1, y0=0, y1=1
)

fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.update_xaxes(constrain='domain')
fig.show()
```

```
# 회귀 평가 RMSE, R2, RMSLE
from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error
import numpy as np
def RMSE(y_test, y_pred):
  return np.sqrt(mean_squared_error(y_test, y_pred))
def RMSLE(y_test, y_pred):
  return np.sqrt(mean_squared_log_error(y_test, y_pred))
rsme = RMSE(y_test, y_pred)
rsmle = RMSLE(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
