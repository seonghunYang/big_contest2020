출루율 => 안타와 사구를 통한 출루 확률 <br/>
장타율 => 장타의 영향력을 확인

```
# 출루율/ 장타력 함수
def runBaseP(row):
    previous = hitter[(hitter['시즌'] == row['시즌']) & (hitter['선수코드'] == row['선수코드']) & (hitter['일자'] < row['일자'])]
    if len(previous) == 0:
        row['출루율'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['안타'] + previous_sum['4구']  + previous_sum['사구']
    down = previous_sum['사구'] + previous_sum['타수'] + previous_sum['4구'] + previous_sum['희비']
    if down == 0:
        row['출루율'] = 0
        return row
    row['출루율'] = up / down
    return row

def longHitP(row):
    previous = hitter[(hitter['시즌'] == row['시즌']) & (hitter['선수코드'] == row['선수코드']) & (hitter['일자'] < row['일자'])]
    if len(previous) == 0:
        row['장타력'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['단타'] * 1 + previous_sum['2루타'] * 2 + previous_sum['3루타'] * 3 + previous_sum['홈런'] * 4
    down = previous_sum['타수']
    if down == 0:
        row['장타력'] = 0
        return row
    row['장타력'] = up / down
    return row
```

```
# 출루율 구하기
# for hitter_i in hitter_list:
#   hitter_i = hitter_i.apply(runBaseP, axis=1)
hitter = hitter.apply(runBaseP, axis=1)
```

타석수가 '시즌 경기수 x 3.1'[3]을 넘겨야 정식 기록으로 인정받는다 타율에 대한 이상치 고려 필요

```
def hitP(row):
    previous = saver[(saver['시즌'] == row['시즌']) & (saver['선수코드'] == row['선수코드']) & (saver['일자'] < row['일자'])]
    if len(previous) == 0:
        row['타율'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['안타']
    down = previous_sum['타수']
    if down == 0:
        row['타율'] = 0
        return row
    row['타율'] = up / down
    return row
```

```
saver = saver.apply(hitP, axis=1)
```

BABIP 구하기 (타자) => 인프레이 타율, 홈런을 제외하고 계산, 상당 부분이 운에 의해서 결정되기 때문에, 운 요소를 파악 가능  삼진을 제외하기 때문에 삼진에 대한 영향력도 따로 구해야한다.

```
def createBABIP(row):
    previous = saver[(saver['시즌'] == row['시즌']) & (saver['선수코드'] == row['선수코드']) & (saver['일자'] < row['일자'])]
    if len(previous) == 0:
        row['BABIP'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['안타'] - previous_sum['홈런']
    down = previous_sum['타수'] - previous_sum['홈런'] - previous_sum['삼진'] + previous_sum['희비']
    if down == 0:
        row['BABIP'] = 0
        return row
    row['BABIP'] = up / down
    return row
```

```
saver = saver.apply(createBABIP, axis=1)
```
