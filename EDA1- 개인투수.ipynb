## 5.평균자책점

```python
pitcher_era = pitcher[['일자', '선수코드', '자책', '이닝*3']]
def createERA(row):
    pitcher_player = pitcher[pitcher['선수코드'] == row['선수코드']].copy()
    pitcher_player = pitcher_player[(pitcher_player['시즌'] == row['시즌']) &(pitcher_player['일자'] < row['일자'])]
    r = pitcher_player['자책'].sum()
    inn = pitcher_player['이닝*3'].sum() / 3
    if inn == 0:
        return 0
    ERA = (r * 9) / inn
    return ERA
pitcher['평균자책점_경기전'] = pitcher.apply(createERA, axis=1)
```
