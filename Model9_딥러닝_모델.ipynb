```
team_code = ['OB', 'WO', 'HT', "LG", 'SS', 'KT', 'LT', 'SK', "HH"]
def team_encoding(row, team_code):
  if row['팀코드'] == team_code:
    row[team_code] = 1
  else:
    if row['팀코드'] != "NC":
      row[team_code] = 0
  return row

def vs_team_encoding(row, team_code):
  if row['상대팀코드'] == team_code:
    row["상대팀{}".format(team_code)] = 1
  else:
    if row['상대팀코드'] != "NC":
      row["상대팀{}".format(team_code)] = 0
  return row


def createHome(row):
  if row['초말'] == 'B':
    row['초말'] = 1
  else:
    row['초말'] = 0
  return row


# 2016 = 0  0 0 0 0 0 0 
year_code = [2017, 2018, 2019, 2020]
def year_encoding(row, year_code):
  if row['시즌'] == year_code:
    row[str(year_code)] = 1
  else:
    if row['시즌'] != 2016:
      row[str(year_code)] = 0
  return row
for team in team_code:
  vs_r_team_hitter = vs_r_team_hitter.apply(team_encoding, axis=1, team_code=team)
  vs_r_team_hitter = vs_r_team_hitter.apply(vs_team_encoding, axis=1, team_code=team)
   
vs_r_team_hitter = vs_r_team_hitter.apply(createHome, axis=1)
for year in year_code:
  vs_r_team_hitter = vs_r_team_hitter.apply(year_encoding, axis=1, year_code=year)

vs_r_team_hitter = vs_r_team_hitter.fillna(0)
```

```
def previousVsData(row, preGame):
  rowData = data[(data['시즌'] == row['시즌']) & (data['팀코드'] == row['팀코드']) & (data['일자'] < row['일자']) & (data['상대팀코드'] == row['상대팀코드'])]
  if len(rowData) <= 2:
    for col in cols:
      row['{}경기전{}'.format(preGame, col)] = 0   
    return row
  rowData = rowData.iloc[-1 * preGame]
  for col in cols:
    row['{}경기전{}'.format(preGame, col)] = rowData[col]
  return row
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size = 0.2 ,shuffle=False
)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics

model = Sequential()
model.add(Dense(128, input_shape=(133, ), activation='relu'))
model.add(Dropout(0.5))  
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.5))  
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.5))  
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.5))  
model.add(Dense(1, activation= 'sigmoid'))

from keras import optimizers

adam = optimizers.Adam(lr=0.005)
model.compile(loss='binary_crossentropy', optimizer = adam, metrics=[metrics.binary_accuracy])

from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=20)
model.fit(x_train, y_train, validation_split=0.2, verbose = 1,
          batch_size=1, epochs=100, callbacks = [early_stopping])

loss, accuracy = model.evaluate(x_test, y_test, batch_size=1)
print('lose: ', loss)
print('mse: ', accuracy)

y_pred = model.predict(x_test)
```
