```
def createERA(row):
    previous = vs_r_team_pitcher[vs_r_team_pitcher['팀코드'] == row['팀코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) &(previous['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
      row['상대최근평균자책점'] = 0
      return row
    up = previous['자책점'].sum()
    down = previous['이닝*3'].sum() / 3
    if down == 0:
        row['상대최근평균자책점']
        return row
    row['상대최근평균자책점'] = (up * 9) / down
    return row

def createRA(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['팀코드'] == row['팀코드'])].copy()
    previous = previous[(previous['시즌'] == row['시즌'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) &(previous['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
      row['상대최근RA/9'] = 0
      return row
    up = previous['실점'].sum()
    down = previous['이닝*3'].sum() / 3
    if down == 0:
        row['상대최근RA/9']
        return row
    row['상대최근RA/9'] = (up * 9) / down
    return row

def createERC(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['팀코드'] == row['팀코드'])].copy()
    previous = previous[(previous['시즌'] == row['시즌'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) &(previous['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
      row['상대최근ERC'] = 0
      return row
    previous_sum = previous.sum()
    PTB = 0.89 * (1.255 * (previous_sum['안타'] - previous_sum['홈런']) + 4 * previous_sum['홈런']) + 0.475 * (previous_sum['4구'] + previous_sum['사구'] - previous_sum['고4'])
    ERC = 9* ( ((previous_sum['안타'] + previous_sum['4구'] + previous_sum['사구']) * PTB) / (previous_sum['타자'] * previous_sum['이닝*3'] / 3) ) - 0.56
    if ERC < 2.24:
      ERC = 9* ( ((previous_sum['안타'] + previous_sum['4구'] + previous_sum['사구']) * PTB) / (previous_sum['타자'] * previous_sum['이닝*3'] / 3) ) * 0.75
    row['상대최근ERC'] = ERC
    return row

def createFIP(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근FIP'] = 0
        return row
    previous_sum = previous.sum()
    FIP = (( 13 * previous_sum['홈런'] + 3 * (previous_sum['4구'] + previous_sum['사구'] - previous_sum['고4']) - 2 * previous_sum['삼진']) / (previous_sum['이닝*3'] / 3)) + C_dic[row['시즌']]
    row['상대최근FIP'] = FIP
    return row

def createLOB(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근LOB'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['안타'] + previous_sum['4구'] + previous_sum['사구'] - previous_sum['실점']
    down = previous_sum['안타'] + previous_sum['4구'] + previous_sum['사구'] - (1.4 * previous_sum['홈런'])
    if down == 0:
      row['상대최근LOB'] = 0
      return row 
    row['상대최근LOB'] = up / down
    return row

def createK_HR(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근홈런율'] = 0
        row['상대최근삼진율'] = 0
        return row
    previous_sum = previous.sum()
    up_HR = previous_sum['홈런'] 
    up_k = previous_sum['삼진']
    down = previous_sum['타자'] - previous_sum['고4']
    if down == 0:
        row['상대최근홈런율'] = 0
        row['상대최근삼진율'] = 0
        return row
    row['상대최근홈런율'] = up_HR / down
    row['상대최근삼진율'] = up_k / down
    return row

def create9(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근K9'] = 0
        row['상대최근BB9'] = 0
        row['상대최근HR9'] = 0
        row['상대최근KBB'] = 0
        return row
    previous_sum = previous.sum()
    up_K = previous_sum['삼진'] 
    up_BB = previous_sum['4구'] - previous_sum['고4']
    up_HR = previous_sum['홈런']
    down = previous_sum['이닝*3'] / 3
    if down == 0 or up_BB == 0 :
        row['상대최근K9'] = 0
        row['상대최근BB9'] = 0
        row['상대최근HR9'] = 0
        row['상대최근KBB'] = 0
        return row
    row['상대최근K9'] = up_K / down * 9
    row['상대최근BB9'] = up_BB / down * 9
    row['상대최근HR9'] = up_HR / down * 9
    row['상대최근KBB'] = up_K / up_BB
    return row

def createWhip(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근whip'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['안타'] + previous_sum['4구']
    down = previous_sum['이닝*3'] / 3
    if down == 0:
      row['상대최근whip'] = 0
      return row 
    row['상대최근whip'] = up / down
    return row

def createPFR(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근PFR'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = (previous_sum["삼진"]+previous_sum["4구"])/(previous_sum["이닝*3"]/3)
    except ZeroDivisionError:
        pg = 0
    row["상대최근PFR"] = pg
    return row
def createPG(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근P/G'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = previous_sum["투구수"]/previous.count()[0]
    except ZeroDivisionError:
        pg = 0
    row["상대최근P/G"] = pg
    return row

def createPIP(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근P/IP'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = previous_sum["투구수"]/(previous_sum["이닝*3"]/3)
    except ZeroDivisionError:
        pg = 0
    row["상대최근P/IP"] = pg
    return row
def createPPA(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근P/PA'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = previous_sum["투구수"]/previous_sum["타자"]
    except ZeroDivisionError:
        pg = 0
    row["상대최근P/PA"] = pg
    return row

def createCYP(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근CYP'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = (previous_sum["이닝*3"]/6-previous_sum["자책점"])+previous_sum["삼진"]/10+len(previous[previous['결과'] == 'W'])
    except ZeroDivisionError:
        pg = 0
    row["상대최근CYP"] = pg
    return row

def createFE(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근F-E'] = 0
        return row
    previous_mean = previous.mean()
    try:
        fe = previous_mean["FIP"]-previous_mean["평균자책점"]
    except ZeroDivisionError:
        fe = 0
    row["상대최근F-E"] = fe
    return row

def createRE(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드'])& (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근R-E'] = 0
        return row
    previous_sum = previous.mean()
    try:
        fe = previous_sum["RA/9"]-previous_sum["평균자책점"]
    except ZeroDivisionError:
        fe = 0
    row["상대최근R-E"] = fe
    return row
```

```
def createkwERA(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근kwERA'] = 0
        return row
    previous_sum = previous.sum()
    try:
        up = 5.40-12*(previous_sum['삼진']-previous_sum['4구'])/previous_sum['타자']
    except ZeroDivisionError:
        row['상대최근kwERA'] = 0
        return row
    row['상대최근kwERA'] = up
    return row

def createK(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근K%'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['삼진']
    down = previous_sum['타자']
    if down == 0:
        row['상대최근K%'] = 0
        return row
    row['상대최근K%'] = up/down
    return row

def createBB(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근BB%'] = 0
        return row
    previous_sum = previous.sum()
    up = previous_sum['4구']
    down = previous_sum['타자']
    if down == 0:
        row['상대최근BB%'] = 0
        return row
    row['상대최근BB%'] = up/down
    return row

def createOAVG(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근OAVG'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oavg = previous_sum["안타"]/previous_sum["타수"]
    except ZeroDivisionError:
        oavg = 0
    row['상대최근OAVG']= oavg
    return row

def createOOBP(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근OOBP'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oobp = (previous_sum["안타"]+previous_sum["4구"]+previous_sum["사구"])/(previous_sum["타수"]+previous_sum["4구"]+previous_sum["사구"]+previous_sum["희비"])
    except ZeroDivisionError:
        oobp = 0
    row['상대최근OOBP']= oobp
    return row

def createOSLG(row):
    previous = vs_r_team_pitcher[(vs_r_team_pitcher['시즌'] == row['시즌']) & (vs_r_team_pitcher['팀코드'] == row['팀코드']) & (vs_r_team_pitcher['상대팀코드'] == row['상대팀코드']) & (vs_r_team_pitcher['일자'] < row['일자'])].tail(3)
    if len(previous) == 0:
        row['상대최근OSLG'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oslg = (previous_sum["단타"]+2*previous_sum["2루타"]+3*previous_sum["3루타"]+4*previous_sum["홈런"])/previous_sum["타수"]
    except ZeroDivisionError:
        oslg = 0
    row['상대최근OSLG']= oslg
    return row
```
