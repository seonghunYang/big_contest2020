```
team_pitcher = pd.read_csv('./gdrive/My Drive/빅콘테스트2020/EDA결과/EDA8-상대전적_투수_세이버 매트릭스_합본.csv', encoding='utf-8-sig')
```

```
import plotly.graph_objects as go

data = team_pitcher[(team_pitcher['팀코드'] == 'LG') & (team_pitcher['상대팀코드'] == 'HH') & (team_pitcher['시즌'] == 2018)].copy()
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['평균자책점'], mode='lines', name='ERA'
    )
)
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['RA/9'], mode='lines', name='RA/9'
    )
)
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['ERC'], mode='lines', name='ERC'
    )
)
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['FIP'], mode='lines', name='FIP'
    )
)
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['K9'], mode='lines', name='K9'
    )
)
fig.add_trace(
    go.Scatter(
        x=data['일자'], y=data['BB9'], mode='lines', name='BB9'
    )
)
fig.update_layout(
  {
      "title": {
          "x": 0.5,
          "y": 0.9,
          "font": {
              "size": 15
          }
      },
      "showlegend": True,
      "xaxis": {
          "title": "일자"
        },
    }
)
fig.show()
```
